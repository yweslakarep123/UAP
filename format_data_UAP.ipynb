```python
from sklearn.model_selection import train_test_split


train_data_A1, test_data_A1 = train_test_split(
    RK_A1,
    test_size=0.3,
    stratify=RK_A1['MotorCondition'],
    random_state=42
)

train_data_A2, test_data_A2 = train_test_split(
    RK_A2,
    test_size=0.3,
    stratify=RK_A2['MotorCondition'],
    random_state=42
)

train_data_A3, test_data_A3 = train_test_split(
    RK_A3,
    test_size=0.3,
    stratify=RK_A3['MotorCondition'],
    random_state=42
)

train_data_A1.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A1/train_data_A1.csv', index=False)
test_data_A1.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A1/test_data_A1.csv', index=False)

train_data_A2.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A2/train_data_A2.csv', index=False)
test_data_A2.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A2/test_data_A2.csv', index=False)

train_data_A3.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A3/train_data_A3.csv', index=False)
test_data_A3.to_csv('/content/drive/MyDrive/Test_dataset/UAP_mlp/RK-A3/test_data_A3.csv', index=False)
```
